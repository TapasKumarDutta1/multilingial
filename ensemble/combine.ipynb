```
import pandas as pd
tst=pd.read_csv( '/content/gdrive/My Drive/multilingual/test.csv.zip')
```

```
import seaborn as sns
import glob
import numpy as np
from matplotlib import pyplot as plt
for path in glob.glob('/content/gdrive/My Drive/multilingual/*_predictions_*.npy'):
    print(path.split('_')[0].split('/')[-1])
    print(path.split('_')[-1].split('.')[0])
    a=np.load(path,allow_pickle=True)
    ls=[]
    for i in a[0]:
      ls.extend(i)
    b=np.vstack(ls)
    print(np.var(b))
    sns.distplot(b)
    plt.show()
```

```
dk={'fr':'french',
    'ru':'russian',
    'it':'italian',
    'pt':'portuguese',
    'es':'spanish',
    'tr':'turkish'}
for en1,lang in enumerate(tst['lang'].unique()):
  df=tst.loc[tst['lang']==lang]
  for en,path in enumerate(glob.glob('/content/gdrive/My Drive/multilingual/'+dk[lang]+'_predictions_*')):
    a=np.load(path,allow_pickle=True)
    try:
      ls=[]
      for i in a[0]:
        ls.extend(i)
      b=np.vstack(ls)
      if np.var(b)>1e-3:
        df['pre_'+str(en)]=b
      else:
        print(path)
    except:
      if np.var(a)>1e-3:
        df['pre_'+str(en)]=a
      else:
        print(path)
    
  if en1==0:
      sudo=df
  else:
      sudo=pd.concat([sudo,df],0)
```

```
sudo['toxic']=sudo[['pre_0','pre_1','pre_2','pre_3','pre_4']].mean(skipna=True,axis=1)
```

```
sudo=sudo[['id','toxic']].sort_values(by='id')
sudo.to_csv('submission.csv')
```
